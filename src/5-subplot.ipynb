### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y_line = [1, 4, 9, 16, 25]      
categories = ['A', 'B', 'C', 'D']  
values_bar = [5, 7, 3, 9]
y_scatter = [2, 4, 6, 8, 10]       
pie_labels = ['Python', 'Java', 'C++', 'JS']  
pie_sizes = [40, 25, 20, 15]


fig, axs = plt.subplots(2, 2, figsize=(10, 8))


axs[0, 0].plot(x, y_line)
axs[0, 0].set_title("Line Plot")


axs[0, 1].bar(categories, values_bar, color=['red', 'blue', 'green', 'orange'])
axs[0, 1].set_title("Bar Chart")


axs[1, 0].scatter(x, y_scatter)
axs[1, 0].set_title("Scatter Plot")


axs[1, 1].pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%')
axs[1, 1].set_title("Pie Chart")


plt.tight_layout()


plt.show()

```
